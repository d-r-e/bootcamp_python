```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

class ColorFilter:
    
    @staticmethod
    def invert(array):
        array =  255 - array
        return array
    
    def test_blue(array):
        return array * np.zeros(np.shape(array))
    def to_blue(array):
        array[..., 0] = 0 #[r g b] r = 0
        array[..., 1] = 0 #[r g b] g = 1
        return array
    
    def to_green(array):
        array[..., 0] = 0
        array[..., 2] = 0
        return array
    
    def to_red(array):
        array[..., 1] = 0
        array[..., 2] = 0
        return array

    def celluloid(array):
        for i in np.shape(array[-1]):
            if i < 128:
                array[..., i] = 0
        return array
```
